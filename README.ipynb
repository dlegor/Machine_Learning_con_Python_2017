## Entorno.


Para seguir los ejemplos del tutorial se necesita contar con los paquetes siguientes:

* [NumPy](http://www.numpy.org/)
* [SciPy](https://www.scipy.org/)
* [matplotlib](http://matplotlib.org/)
* [seaborn](https://seaborn.pydata.org/)
* [Pandas](http://pandas.pydata.org/)
* [IPython](https://ipython.org/)
* [Jupyter Notebook](http://jupyter.org/)
* [Scikit-Learn](http://scikit-learn.org/stable/)
* [StatsModels](http://www.statsmodels.org/stable/index.html)
* [PyYaml](http://pyyaml.org/)
  
*Nota*: Todos los paquetes se encuentran disponibles en la distribución Python de [Anaconda.](https://www.continuum.io/)  

Para validar que se cuenta con lo necesario se ejecuta el script *entorno.py*, se puede hacer de las siguientes formas:

* Desde consola
* Desde un Notebook

Desde *consola* basta con ubicarnos en el directorio correspondiente al material, identificar el script con el nombre *entorno.py* y ejecutar la siguiente instrucción en la línea de comandos. Las instrucciones serían las siguientes:

```bash

cd /Directorio/donde/se/ubica/el/material

#Ubicado en el directorio Notas_Curso_ML_2017
python entorno.py

``` 
La salida en la terminal se mostrará de la siguiente forma:
![](Imagenes/Pantalla_salida_entorno.png)

Desde un *Notebook* se debe primero ejecutar en la linea de comandos desde la ruta o posición que se desee (de preferencia en la ruta del material) lo siguiente:

``` bash
jupyter notebook

```

Esto abrirá un servicio en el browser que se tenga asignado como predeterminado, de preferencia abrilo en Google Chrome o Firefox. En la pestaña del browser se visualizará lo siguiente:

![Detalles_Jupyter](Imagenes/Detalles_Jupyter_Notebook.png)

Haz doble click o abre el notebook que tiene por nombre **Validacion_Entorno.ipynb**, este cuenta con dos celdas que tienen las siguente instrucciones:

``` bash

%run entorno.py

#Se encuentran en celdas separadas

%run instalaciones_faltantes.py

```
Para ejecutar solo se tienen que presionar en el botón de *run cell* para cada celda seleccionada, en la siguiente imagen se muestra:

![](Imagenes/Una_celda.png)


Para ejecutar todas las celdas se elige la opcion en el *Menú-> cell->Run All*, la siguiente imagen muestra como ejecutarlo:

![](Imagenes/Celds_Todas.png)


## Datos del Tutorial


Si descargas las notas del taller desde el repositorio en Github contarás con ellos, en todo caso si por alguna razón no se cuentan con ellos se pueden descargar con el script *datos_taller.py*. Estos crearán las carpetas necesarias dentro del directorio *DataSets*.

Las instrucciones de ejecución son similares a las realizadas para validar el entorno, tanto para la ejecución en consola como en un notebook.

## Lista de referencias.

En cada uno de los Notebook se tienen algunas referencias, tanto artículos de investigación como capítulos de libros donde se explican a detalle los temas mencionados.

## Agradecimietos.

El material del taller no sería posible sin toda la documentación disponible en las páginas de las bibliotecas como [Pandas](http://pandas.pydata.org/), [Numpy](http://www.numpy.org/), [SciPy](https://www.scipy.org/),[Scikit-learn](http://scikit-learn.org/stable/),etc. Buena parte de la documentación se le debe a la comunidad Open Source. Buena parte del material no sería posible sin todos los tutoriales, libros y material compartido en la web por [Wes McKinnney](http://wesmckinney.com/), [Kyle Kastner](https://kastnerkyle.github.io/), [Jhon Miles White](http://www.johnmyleswhite.com/), [Andreas Muller](http://amueller.github.io/), [Sebastina Raschka](https://sebastianraschka.com/),[Peter Norvig](http://norvig.com/), entre tantos otros.
