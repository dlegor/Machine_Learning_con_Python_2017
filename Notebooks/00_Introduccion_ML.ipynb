### Ejemplos en Python 

Como ejemplo se resuelve el problema de las **Torres de Hanoi**, para ejecutar el código solo basta elegir el botón *run cell* o combinar las teclas **Ctrl+Enter** al colocarse sobre la celda donde se encuentra el código.

```python
#Código de Brad Miller and David Ranum


#Definicion de la Funcion de Movimientos
def moveTower(height,fromPole, toPole, withPole):
    if height >= 1:
        moveTower(height-1,fromPole,withPole,toPole)
        moveDisk(fromPole,toPole)
        moveTower(height-1,withPole,toPole,fromPole)

#Presentación de los Movimientos realizados        
def moveDisk(fp,tp):
    print("Moviendo el disco desde la torre",fp,"a la torre",tp)

#Ejecución de programa
moveTower(3,"A","B","C")
```

El ejemplo anterios muestra como definir una función y hacer uso de la recursividad, con un poco de esfuezo se puede ir leyendo las indicaciones y resulta muy fácil de leer el código. Suele hacerse el comentario en la programación en Python: *"Escribir código en Python es casi como escribir seudo código"*.

El siguiente ejemplo muestra como se obtiene un mismo resultado de diferentes maneras, la intención es mostrar la diversidad de formas  y no tanto ver cuál es la mejor manera de construir una lista.

```python
# Se genera una rango de números
Rango=range(0,10)

#Se define la lista vacía
L=[]

#Ciclo sobre los elementos de Rango
for i in Rango:
    L.append(i*i)#Se van agregando los elementos de Rango

#Se presenta la lista    
print(L)    
```

```python
#Construcción de una lista con los elementos de Rango

#Listas por compresión
[i*i for i in Rango]
```

```python
#Se contruye una lista con todos los números, pero se hace uso de una función lambda, útiles en
# programación funcional

list(map(lambda x: x*x,Rango))        
```

Los 3 ejemplo muestran aspectos diferentes de la programación en Python. En el primero ejemplo se hace uso del método *append* de las listas. Lo que hace es agregar al final cada nuevo elemento.

Para el segundo ejemplo se hace uso de una *lista por compresión*, es un modo de crear listas desde otra lista (también se puede hacer desde los elementos de un diccionario). Este tipo de construcción es más similar a la notación de teoría de conjuntos. Si se observa la notación de la lista y la definición del siguiente conjunto podrás observar la similaridad.

\begin{align}
Conjunto=\{x*x \mid \text{ Para todo x en Rango}\}
\end{align}

En el último ejemplo, se usan las funciones *list, map y lambda*. La función *list()* es como su nombre lo indica para contruir una lista. La función *map()* es un mapeo de la función lambda sobre los elementos que están en Rango. Esta es una función estandar en programación funcional y funciona igual que el concepto de función en un sentido matemático.

La función *lamnda* es un tipo de función estandar en la programación funcional, se les llama *funciones anónimas o expresiones lambda*. Ayudan a definir funciones, qué son aplicadas al momento, pero no reusables. Ejemplo el último codigo se podía programar contando con la definición normal de una función y aplicarla con *map()*.

~~~ python
def cuadrado(x): return x*x

list(map(cuadrado,Rango))
~~~

El resultado es el mismo que se obtiene con el uso de la función lambda. Para ver algunos ejemplos de programación funcional básica con Python se pueden consultar en la siguientes referencias:

*Como ejemplo se hacen algunas operaciones básicas sobre matrices haciendo uso de NumPy y con SciPy se hace un ejemplo calculando algunos estadísticos.*

```python
#Ejemplo de uso de NumPy

#Se carga el módulo
import numpy as np

#Se definen 2 array en NumPy
x = np.array([[1,2],[3,4]], dtype=np.float64)
y = np.array([[5,6],[7,8]], dtype=np.float64)

#Se realizan operaciones "Vectorizadas".

# Elemento a elemento se suman; 
# [[ 6.0  8.0]
#  [10.0 12.0]]
print("Suma elemento a elemento\n")
print(x + y)
print()
print(np.add(x, y))
print()
print()

# Elemendo a elemento se restan
# [[-4.0 -4.0]
#  [-4.0 -4.0]]
print("Resta Elemento a Elemento\n")
print(x - y)
print()
print(np.subtract(x, y))
print()
print()


# Se hace el producto elemento a elemento
# [[ 5.0 12.0]
#  [21.0 32.0]]
print("Multiplicación Elemeto a Elemento\n")
print(x * y)
print()
print(np.multiply(x, y))
print()
print()

# Se hace la división elemento a elemento
# [[ 0.2         0.33333333]
#  [ 0.42857143  0.5       ]]
print("División\n")
print(x / y)
print()
print(np.divide(x, y))
print()
print()

# Se hace la división elemento a elemento
# [[ 1.          1.41421356]
#  [ 1.73205081  2.        ]]
print("Raiz Cuadrada de cada Elemento\n")
print(np.sqrt(x))
```

El beneficio de usar NumPy es poder hacer operaciones vectorizadas sobre arreglos o array de datos, además de otras funcionalidades que se tienen para el procesamiento de datos, las cuales son aprovechadas por **Pandas**.

```python
#Ejemplo de SciPy

from scipy import stats

#Generacion de Datos Aleatorios
np.random.seed(282629734)
x = stats.t.rvs(10, size=1000)

print('Estadísticos Básicos desde los Array de NumPy')
print("Max   y  Min")
print(x.max(), x.min())  # es equivalente a calcular np.max(x), np.min(x)
print(' Media   y  Varianza')
print(x.mean(), x.var()) # es equivalente a np.mean((x), np.var(x)
print()
#Se puede hacer asignación multiple en Python solo separado por ","

#Se toman valores de una distribución T-Student
m, v, s, k = stats.t.stats(10, moments='mvsk')

#Se asigna los estadisticos básicos
n, (smin, smax), sm, sv, ss, sk = stats.describe(x)
print('Estadísticos con SciPy')
print('Distribución:')
#Distribución:
sstr = 'media = %6.4f, varianza = %6.4f, asimetria = %6.4f, kurtosis = %6.4f'
print(sstr %(m, v, s ,k))
print()
print('Muestra:      ')
print(sstr %(sm, sv, ss, sk))
```
