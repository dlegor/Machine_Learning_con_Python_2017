## Ejemplo Sencillo

Se hace una regresión lineal como ejemplo.

```python
#Codigo de la regresión lineal


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression #Regresion Lineal


#configuraciones de tipo de gráficos y de tamaño
plt.rcParams["figure.figsize"] = (12, 5)
plt.rcParams['figure.titlesize'] = 'large'
plt.rcParams['font.size'] = 12
plt.rcParams['axes.titlesize']=15

#Se generan los datos 
x=np.random.normal(size=1000)
y=np.random.normal(size=1000)+2.5*x

#Se define el modelo
Modelo=LinearRegression()
Modelo.fit(x.reshape((-1,1)),y)

#Se realiza una predicción
Pred=Modelo.predict(x.reshape((-1,1)))

#Se grafican los datos
plt.plot(x,y,'ko')
plt.plot(x,Pred,'r-')
plt.legend(['Datos Originales','Recta Estimada'])
plt.title("Regresion Lineal Simple")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
```

Para el ejemplo anterior se generan datos aleatorios que son definidos como variables $x$ y $y$, intencionalmente los datos son generados desde una *distribución normal*. La relación entre las variables es: $$y=2.5x+\epsilon$$

Donde $\epsilon \in N(0,1)$, donde con $N$ indica que es una distribución normal con media cero y varianza $1$. Solo se realizó la estimación de la recta mediante el modelo `LinearRegression()`. Más adelante se explican detalles de los modelos lineales.
